```python
# load tmdb-movies csv into a dataframe
movies_df = pd.read_csv('tmdb-movies.csv')
```

```python
# remove duplicated
movies_df.drop_duplicates(inplace=True)
```

```python
# check shape of the dataframe again
movies_df.shape
```

```python
# drop imdb_id, homepage, tagline and overview columns. 
movies_df.drop(['imdb_id', 'homepage','tagline', 'overview'], axis=1, inplace=True )
movies_df.shape
```

```python
# remove rows if the budget or revenue or budget_adj or revenue_adj is 0.0
zero_columns = ['budget', 'revenue', 'budget_adj', 'revenue_adj']
movies_df = movies_df.loc[(movies_df.budget * movies_df.revenue * movies_df.budget_adj* movies_df.revenue_adj) != 0]
movies_df.shape
```

```python
# fill the missing values in the following columns with the values indicated
# cast = 'no_cast'
# director = 'no_director'
# keywords = 'no_keywords'
# genres = 'no_genres'
# production_companies = 'no_production_companies'

columns_to_fillna = ['cast', 'director', 'keywords', 'genres', 'production_companies']
for column in columns_to_fillna:
    movies_df[column] = movies_df[column].fillna('no_'+column)

# check the number of null values again
movies_df.isnull().sum()
```

```python
# group all by the year released and plot the popular genre per year
popular_genre_per_year = movies_df.groupby(['release_year'])[['vote_average','genres']].max()
plt.subplots(figsize=(25, 10))
plt.bar(popular_genre_per_year.index, popular_genre_per_year['vote_average'], width=0.9)
plt.xticks(popular_genre_per_year.index, popular_genre_per_year['genres'], rotation='vertical', fontsize=20)
plt.title('The Yearly Most Popular Movie Genre From 1960 to 2015', fontsize=30)
plt.xlabel('Year', fontsize=30)
plt.ylabel('Popularity (Average Votes)', fontsize=30);
```

```python
movie_by_genre = movies_df.groupby(['release_year','genres']).mean()[['popularity', 'vote_average']]
movie_by_genre
```

```python
# Group The Genres by Year and Find the Most Popular Ones Based on popularity and vote_average
for year in np.arange(1960,2016,10): # Interval can be changed if all the years need to appear.
    df = movies_df.query('release_year == @year').groupby('genres').mean().sort_values(by=['vote_average'], ascending=False).head(5)
    plt.subplots(figsize=(25, 20))
    plt.bar(df.index,df['vote_average'])
    plt.xticks(rotation='vertical', fontsize=20)
    plt.title('5 Most Popular Genres Each Year', fontsize=40)
    plt.xlabel(year, fontsize=30)
    plt.ylabel('Popularity', fontsize=30)
```

```python
# yearly revenue change for all movies
yearly_revenue_change = movies_df.groupby('release_year').sum()
yearly_revenue_change.apply(lambda x:x.diff().fillna(0))[['revenue']].plot();
```

```python
# compare the yearly sum for budget and revenue
yearly_revenue_change[['budget_adj', 'revenue_adj']].plot();
```

```python
# Find how the revenue and runtime of a movie are related
ninety_percentile = np.percentile(movies_df['revenue_adj'], 99)
highest_revenue_movies = movies_df.query('revenue_adj > @ninety_percentile ')
highest_revenue_movies['vote_average'].hist();
```

```python
highest_revenue_movies['popularity'].hist();
```

```python
highest_revenue_movies['runtime'].hist();
```

```python
movies_df.groupby('release_year').revenue_adj.mean().hist();
```

```python
# Find the directors for the top 20 movies with highest revenue
top_20_revenue = highest_revenue_movies.sort_values(by=['revenue_adj'], ascending=False).head(20)
plt.subplots(figsize=(25, 10))
plt.bar(top_20_revenue['director'],top_20_revenue['revenue_adj'])
plt.xticks(top_20_revenue['director'],rotation='vertical', fontsize=20)
plt.title('Movies with Highest Revenues by Director Names', fontsize=40)
plt.xlabel('Director', fontsize=30)
plt.ylabel('Revenue', fontsize=30);
```

```python
# Find the number of times a particular director made it to the top 20 movies
directors = [director.split('|') for director in top_20_revenue.director]
top_directors = []
for director in directors:
    for single in director:
        top_directors.append(single)


director_counts = dict()
for i in top_directors:
    director_counts[i] = director_counts.get(i, 0)+1

director_count_df = pd.DataFrame.from_dict(director_counts, orient="index")
plt.subplots(figsize=(25, 10))
plt.bar(director_count_df.index,director_count_df[0]);
plt.xticks(director_count_df.index,rotation='vertical', fontsize=20)
plt.title('Number of Top Movies Per Director', fontsize=40)
plt.xlabel('Director Name', fontsize=30)
plt.ylabel('Count of Top Movies', fontsize=30);
```

```python
# Find the production companies for the top 20 movies with highest revenue
plt.subplots(figsize=(25, 10))
plt.bar(top_20_revenue['production_companies'],top_20_revenue['revenue_adj'])
plt.xticks(top_20_revenue['production_companies'],rotation='vertical', fontsize=20)
plt.title('Movies with Highest Revenues by Production Company Names', fontsize=40)
plt.xlabel('Production Companies', fontsize=30)
plt.ylabel('Revenue', fontsize=30);
```

```python
# Find the number of times a particular director made it to the top 20 movies
companies = [company.split('|') for company in top_20_revenue['production_companies']]
top_companies = []
for company in companies:
    for single in company:
        top_companies.append(single)

company_counts = dict()
for i in top_companies:
    company_counts[i] = company_counts.get(i, 0) + 1
company_count_df = pd.DataFrame.from_dict(company_counts, orient="index")

plt.subplots(figsize=(25, 10))
plt.bar(company_count_df.index,company_count_df[0]);
plt.xticks(company_count_df.index,rotation='vertical', fontsize=20)
plt.title('Number of Top Movies Per Director', fontsize=40)
plt.xlabel('Director Name', fontsize=30)
plt.ylabel('Count of Top Movies', fontsize=30);
```

```python
# Find the keywords that are associated with the top revenue movies
plt.subplots(figsize=(25, 10))
plt.bar(top_20_revenue['original_title'],top_20_revenue['revenue_adj'])
plt.xticks(top_20_revenue['original_title'],rotation='vertical', fontsize=20)
plt.title('Revenue By Movie Title', fontsize=40)
plt.xlabel('Movie Title', fontsize=30)
plt.ylabel('Revenue', fontsize=30);
```

```python
# Find the keywords that are associated with the top revenue movies
plt.subplots(figsize=(25, 10))
plt.bar(top_20_revenue['keywords'],top_20_revenue['revenue_adj'])
plt.xticks(top_20_revenue['keywords'],rotation='vertical', fontsize=20)
plt.title('Revenue by Keyword', fontsize=40)
plt.xlabel('Keywords', fontsize=30)
plt.ylabel('Revenue', fontsize=30);
```

```python
# Find the number of times a particular keyword made it to the top 20 movies
keywords = [keyword.split('|') for keyword in top_20_revenue['keywords']]
top_keywords = []
for keyword in keywords:
    for single in keyword:
        top_keywords.append(single)

keyword_counts = dict()
for i in top_keywords:
    keyword_counts[i] = keyword_counts.get(i, 0) + 1

keyword_count_df = pd.DataFrame.from_dict(keyword_counts, orient="index").sort_values(by=[0], ascending=False).head(20)

plt.subplots(figsize=(25, 10))
plt.bar(keyword_count_df.index,keyword_count_df[0]);
plt.xticks(keyword_count_df.index,rotation='vertical', fontsize=20)
plt.title('Number of Top Movies Per Director', fontsize=40)
plt.xlabel('Director Name', fontsize=30)
plt.ylabel('Count of Top Movies', fontsize=30);
```

```python
# Find the Actor who is most present in high revenue movies

# Find the number of times a particular keyword made it to the top 20 movies
actors = [actor.split('|') for actor in top_20_revenue['cast']]
top_actors = []
for actor in actors:
    for single in actor:
        top_actors.append(single)

actor_counts = dict()
for i in top_actors:
    actor_counts[i] = actor_counts.get(i, 0) + 1

actor_count_df = pd.DataFrame.from_dict(actor_counts, orient="index").sort_values(by=[0], ascending=False).head(20)

plt.subplots(figsize=(25, 10))
plt.bar(actor_count_df.index,actor_count_df[0]);
plt.xticks(actor_count_df.index,rotation='vertical', fontsize=20)
plt.title('Number of Top Movies Per Director', fontsize=40)
plt.xlabel('Director Name', fontsize=30)
plt.ylabel('Count of Top Movies', fontsize=30);
```
