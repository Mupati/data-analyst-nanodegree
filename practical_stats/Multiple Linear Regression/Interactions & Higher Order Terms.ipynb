```python
import numpy as np
import pandas as pd
import statsmodels.api as sm;

df = pd.read_csv('house_prices.csv')
df.head()
```

```python
df['bedrooms_squared'] = df['bedrooms']*df['bedrooms']
```

```python
df['intercept'] = 1
lm = sm.OLS(df['price'], df[['intercept', 'bedrooms', 'bedrooms_squared']])
results = lm.fit()
results.summary()
```

```python
df['bedrooms_cubed'] = df['bedrooms']*df['bedrooms_squared']
lm = sm.OLS(df['price'], df[['intercept', 'bedrooms', 'bedrooms_squared', 'bedrooms_cubed']])
results = lm.fit()
results.summary()
```

```python
df['area_bed'] = df['area']*df['bedrooms']
lm = sm.OLS(df['price'], df[['intercept', 'area', 'area_bed', 'bedrooms', 'bedrooms_squared', 'bedrooms_cubed']])
results = lm.fit()
results.summary()
```
