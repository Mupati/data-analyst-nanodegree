```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('house_prices.csv')
df.head()
```

```python
df[['A', 'B', 'C']] = pd.get_dummies(df['neighborhood'])
df
```

```python
df[['lodge', 'ranch', 'victorian']] = pd.get_dummies(df['style'])
```

```python
df
```

```python
df['intercept'] = 1
```

```python
lm = sm.OLS(df['price'], df[['intercept', 'lodge', 'ranch']])
res = lm.fit()
res.summary()
```
